### 1. Generating the data

To generate the data, we first generate a random value for W_true and b_true. Then, we take a hundred random points for x and generate y as y = W_true*x + b_true. Lastly, we add some noise to y.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
m = 200 #Traditionally, m is the number of data points
variance_X = 100
variance_Y_noise = 10
W_true = np.random.rand(1) #np.random.rand(d0, d1, d2, ...). Arguments are output dimensions.
b_true = np.random.rand(1)
X = np.random.rand(m)*variance_X
Y = W_true*X + b_true   # The * operator for np arrays is elementwise multiplication.
                        # Use @ for matrix multiplication
Y = Y + np.random.rand(m)*variance_Y_noise
```

Let's have a look at the data.

```python
plt.scatter(X, Y)
plt.show()
```

### 2. Initialize W and b

```python
W = np.random.rand(1)
b = np.random.rand(1)
Y_hat = W*X + b
plt.plot(X, Y_hat, color = "red")
plt.scatter(X, Y)
plt.show()
```

```python
costs = []
alpha = 0.0005
num_iters = 100000
for i in range(num_iters):
    cost = (1/(2*m))*np.sum((W*X + b - Y)**2) #This works because of "broadcasting". Google it
    costs.append(cost)
    deriv_wrt_W = (1/m)*np.sum(X*(W*X + b - Y)) 
    deriv_wrt_b = (1/m)*np.sum(W*X + b - Y)
    W -= alpha*deriv_wrt_W
    b -= alpha*deriv_wrt_b
Y_hat = W*X + b
plt.scatter(X, Y)
plt.plot(X, Y_hat, color = 'red')
plt.show()
plt.plot(costs)
plt.show()
```
